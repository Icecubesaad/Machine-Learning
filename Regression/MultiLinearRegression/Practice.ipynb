```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('SampleData.csv')
X=data.iloc[:,1:].values    # independent stats
Y=data.iloc[:,0].values     # dependent stats
Y=Y.reshape(-1,1)
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)

import statsmodels.api as sm

array_of_once = np.ones((8,1)).astype(int)

X=np.append(arr=array_of_once,values=X,axis=1)

X_pr = np.array(X[:,[0,1,2,3]])

regressor_OLS = sm.OLS(endog=Y,exog=X_pr).fit()
regressor_OLS.summary()
X_pr = np.array(X[:,[0,1,2]])

regressor_OLS = sm.OLS(endog=Y,exog=X_pr).fit()
regressor_OLS.summary()
X_pr = np.array(X[:,[0,2]])

regressor_OLS = sm.OLS(endog=Y,exog=X_pr).fit()
regressor_OLS.summary()
```
