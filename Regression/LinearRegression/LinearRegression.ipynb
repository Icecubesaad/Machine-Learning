```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linear Regression

# data sets

datasets = pd.read_csv('./Salary_Data.csv')
# Y -> dependannt variable  (salary)
# X -> independent variable (experience)

X= datasets.iloc[:,0].values
Y= datasets.iloc[:,-1].values
# Y=[int(y) for y in Y]
# Y
#test and train the data

X = X.reshape(-1, 1)    #tuen in 2d
Y = Y.reshape(-1, 1)    #turn in 2d

from sklearn.model_selection  import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=1/3,random_state=0)
# print(X_train)
# print(X_test)
# print(Y_test)
# print(Y_train)


# finding linear regression



from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)

# predict

y_pred =  regressor.predict(X_test)


# visualized

plt.scatter(X_train,Y_train,c='red')
plt.plot(X_train,regressor.predict(X_train),c='blue')
plt.title('linear regression example of salary and experience')
plt.ylabel('salary')
plt.xlabel('experience')
```
